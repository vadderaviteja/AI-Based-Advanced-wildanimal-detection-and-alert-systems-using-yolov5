```python
%%writefile app.py
import streamlit as st
import torch
import numpy as np
from PIL import Image
import base64
import io

# ----------------------------------------------------
# LOAD YOLOv5 MODEL
# ----------------------------------------------------
model = torch.hub.load(
    'ultralytics/yolov5',
    'custom',
    path='/content/yolov5/runs/train/exp/weights/best.pt',
    force_reload=True
)

# ----------------------------------------------------
# SOUND FUNCTION
# ----------------------------------------------------
def play_alarm_sound():
    alarm_sound = """
    <audio autoplay>
      <source src="https://www.soundjay.com/misc/sounds/bell-ringing-05.mp3" type="audio/mpeg">
    </audio>
    """
    st.markdown(alarm_sound, unsafe_allow_html=True)

# ----------------------------------------------------
# STREAMLIT UI
# ----------------------------------------------------
st.title("ğŸ¾ Wild Animal Detection System")
st.write("Upload an image to detect **Human / Pet Animal / Wild Animal**.")

uploaded_file = st.file_uploader("Upload Image", type=["jpg", "jpeg", "png"])

if uploaded_file:

    # Load the image
    img = Image.open(uploaded_file)
    st.image(img, caption="Uploaded Image", use_column_width=True)

    # Run YOLO prediction
    results = model(img)

    # Convert predictions
    predictions = results.pandas().xyxy[0]

    st.subheader("Detection Results:")
    st.write(predictions)

    # Display processed image with bounding boxes
    result_img = Image.fromarray(results.render()[0])
    st.image(result_img, caption="Detection Output", use_column_width=True)

    # ----------------------------------------------------
    # DECISION LOGIC
    # ----------------------------------------------------
    detected_classes = predictions['name'].tolist()

    if "wild_animal" in detected_classes:
        st.error("âš ï¸ Wild Animal Detected! Triggering Alarm!")
        play_alarm_sound()

    elif "Domestic_animal" in detected_classes:
        st.success("ğŸ¶ Pet Animal Detected!")

    elif "human" in detected_classes:
        st.info("ğŸ§ Human Detected!")

    else:
        st.warning("â“ No known objects detected.")
```
